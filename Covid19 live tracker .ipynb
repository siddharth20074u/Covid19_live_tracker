```python
import pandas as pd

df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df.tail()
```

```python
df = df.iloc[:-1, 1:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

total_states = np.arange(len(df['state_name']))
```

```python
from matplotlib.pyplot import figure

figure(num = None, figsize = (12, 12), dpi = 100, facecolor = 'w', edgecolor = 'k')
plt.barh(total_states, df['positive'], align = 'center', alpha = 0.5, color = (1, 0, 0), edgecolor = (0.5, 0.2, 0.8))
plt.yticks(total_states, df['state_name'])
#plt.xlim(0, max(df['positive'])+100)
plt.xlabel('Positive Number of cases')
plt.title('Corona Virus cases')
plt.show()
```

```python
figure(num = None, figsize = (12, 12), dpi = 100, facecolor = 'w', edgecolor = 'k')
plt.barh(total_states, df['cured'], align = 'center', alpha = 0.5, color = (0, 1, 0), edgecolor = (0.5, 0.3, 0.8))
plt.yticks(total_states, df['state_name'])
#plt.xlim(1, max(df['cured'])+100)
plt.xlabel('Cured Number of cases')
plt.title('Corona Virus cases')
plt.show()
```

```python
figure(num = None, figsize = (8, 6), dpi = 100, facecolor = 'w', edgecolor = 'k')
plt.barh(total_states, df['death'], align = 'center', alpha = 0.5, color = (0, 0, 1), edgecolor = (0.5, 0.3, 0.8))
plt.yticks(total_states, df['state_name'])
plt.xlabel('Death Number of cases')
plt.title('Corona Virus cases')
plt.show()
```

## Stacking the values in bar plot 

```python
df1 = df.set_index('state_name', drop = True)
```

```python
df1.plot.barh(stacked = True, figsize = (12,10))
plt.xlabel('Number of cases')
plt.title('Corona Virus cases')
plt.show()
```

```python
df2 = df1.iloc[:, 1:3]
df2.plot.barh(figsize = (14, 10))
plt.xlabel('Number of cases')
plt.title('Corona Virus cases')
plt.show()
```
